This is an expoloration of data on college graduats from 2010 to 2012.  Through various data visualizations, we will seek to answer the following questions:

1) Do students in more popular majors make more money?
2) Which majors are predominantly male?  Which majors are predominantly female?
3) Which categories of majors have the most students?

```python
import pandas as pd
import matplotlib
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")
```

```python
# Drop rows with 'NA' to facilitate the visualizations
recent_grads = recent_grads.dropna()
recent_grads.info()
```

```python
ax1 = recent_grads.plot(x='Sample_size', y='Employed', kind='scatter')
ax1.set_title('Employed vs. Sample_size')
```

```python
ax2 = recent_grads.plot(x='Sample_size', y='Median', kind='scatter')
ax2.set_title('Employed vs. Median')
```

The visual above indicates that majors from very popular majors (ie over 100k students) do not make more money on average.  Most of the points are towards the bottom of the range for median salaries.  Only a few points top $40k and none are above $60k.  

```python
ax3 = recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter')
ax3.set_title('Sample_size vs. Unemployment_rate')
```

```python
ax4 = recent_grads.plot(x='Full_time', y='Median', kind='scatter')
ax4.set_title('Full_time vs. Median')
```

As seen above, there appears to be a slight negative relationship betweeen majors with high numbers of full time employees and median salaries.  That is majors with high numbers of full time employees (ie >100k) tend to have slightly lower median salaries (in the $30 to $40k range).  On the other hand, majors with smaller numbers of full time employees tend to have slightly higher median salaries.

```python
ax5 = recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter')
ax5.set_title('ShareWomen vs. Unemployment_rate')
```

The visual above indicates that there is no discernible relationship between majors with a high percentage of females and unemployment rates.  Some majors with a high number of females have relatively high unemployment rates while some majors with a high number of females have low unemployment rates.  

```python
ax6 = recent_grads.plot(x='Men', y='Median', kind='scatter')
ax6.set_title('Men vs. Median')
```

```python
ax7 = recent_grads.plot(x='Women', y='Median', kind='scatter')
ax7.set_title('Women vs. Median')
```
